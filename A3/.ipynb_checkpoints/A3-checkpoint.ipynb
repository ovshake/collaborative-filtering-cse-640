```python
users = pd.read_csv('./data/users.csv')
items = pd.read_csv('./data/movies.csv')
print(len(users))
print(len(items))
print(items.head())
users.head()
```

```python
# Changing the user IDs and movie IDs to numerically sequential IDs 
old_movie_id = items['movie_id'].values
new_to_old_movie_id = {i : old_movie_id[i-1] for i in range(1,len(items) + 1)}
old_user_id = users['_id'].values
new_to_old_user_id = {i : old_user_id[i-1] for i in range(1, len(users) + 1)} 
new_to_old_user_id
```

```python
old_to_new_user_id = {v:k for (k,v) in new_to_old_user_id.items()} 
old_to_new_movie_id  = {v:k for (k,v) in new_to_old_movie_id.items()}
len(old_to_new_movie_id) , len(old_to_new_user_id)
```

## Making The User Item Matrix


```python
rating_matrix = np.zeros((924 , 2850)) 
for i in range(data.shape[0]):
    id = data.loc[i , '_id']
    new_id = old_to_new_user_id[id] 
    rating_dict = data.loc[i , 'rated']
    for k in rating_dict:
        if "s" in k:
            continue
        new_movie_id = old_to_new_movie_id[k] 
        rating_matrix[new_id - 1][new_movie_id-1] = int(rating_dict[k][0])
        
rating_matrix    
    
```

```python
np.unique(rating_matrix)
```

#### Number of Genres And Range Of IMDB Ratings And Range Of Release Dates and also Languages Known


```python
genre = set() 
ratings = set()
release_year = set() 
languages = set() 
for i in range(items.shape[0]):
    g = items.loc[i , 'genre']
    for u in g[1:-1].split(','):
        u = u.replace('"', '')
        u = u.replace(" ", '')
        genre.add(u) 
    date = str(items.loc[i,'released']).split("-")[0]
    release_year.add(date) 
    rt = items.loc[i,'rating'] 
    ratings.add(int(rt)) 
release_year.remove('nan')
len(genre) , len(ratings) , len(release_year) , max(release_year) , min(release_year), ratings
```

So, To encode genre we need 6 bits, for ratings we need 3 bits and for release yeear we need 3 bits for (old, middle, and recent years) 

```python
genre_to_code = {i : g for i,g in enumerate(genre)} 
code_to_genre = {v : k for (k,v) in genre_to_code.items()} 
code_to_genre
```

Now Lets look at the users and what all we can use. 
We can use languages, the DOB the gender and the job 

```python
users.head() 
```

```python
dobs = set() 
gender = set() 
jobs = set() 
languages = set() 
for i in range(users.shape[0]):
    year = str(users.loc[i , 'dob']).split('-')[-1]
    dobs.add(year) 
    gender.add(users.loc[i , 'gender']) 
    jobs.add(users.loc[i,'job'])
    l = users.loc[i , 'languages']
#     print(l)
    if type(l) != float:
        l = l[1:-1] 
        l = l.split(',')
        for v in l:
            v = v.replace('"', '')
            v = v.replace(" ", '')
            languages.add(v)
dobs.remove('nan') 
dobs.remove('')
len(dobs) , len(gender) , len(jobs) , len(languages), max(dobs) , min(dobs)
```

We will be considering those born after 1990 as modern, between 1970 and 1990 as medieval and the rest as old. Hence a 3 bit vector 


```python
code_to_job = {i : v for i,v in enumerate(jobs)}
job_to_code = {v : k for (k,v) in code_to_job.items()} 
job_to_code
```

```python
code_to_lang = {i : v for i,v in enumerate(languages)}
lang_to_code = {v : k for (k,v) in code_to_lang.items()}
lang_to_code
```

```python
code_to_gender = {i : v for i,v in enumerate(gender)}
gender_to_code = {v : k for (k,v) in code_to_gender.items()}
gender_to_code
```
